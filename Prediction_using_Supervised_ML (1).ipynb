```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data= "http://bit.ly/w-data"
df= pd.read_csv(data)
df.head(10)
```

```python
df.describe
```

# Data Visualisation

```python
plt.scatter(df['Hours'],df['Scores'],marker='*')
plt.title('Hours vs Scores')
plt.xlabel("Hours studied")
plt.ylabel('Score obtained ')
plt.show()
```

```python
plt.plot(df['Hours'],df['Scores'],marker='o')
plt.title('Hours vs Scores')
plt.xlabel("Hours studied")
plt.ylabel('Score obtained ')
plt.show()
```

# Train and Test set

```python
X = df.iloc[:, :-1].values  
Y = df.iloc[:, 1].values 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.25,random_state=30)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
plt.scatter(X,Y)
plt.xlabel('Train set')
plt.ylabel('Test set')
plt.show()
```

```python
sns.regplot(x=X,y=Y)
plt.xlabel('Train set')
plt.ylabel('Test set')
```

# Prediction

```python
print(x_test)
y_pred= LR.predict(x_test)
```

```python
df_1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df_1 
```

```python
hours = pd.array(df['Hours'])
own_pred = LR.predict(hours[15].reshape(1,-1))
print("No of Hours = {}".format(hours[15]))
print("Predicted Score = {}".format(own_pred[0]))
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
